```python
## Insertion Sort
def insertionSort(array, count):
    for step in range(1, len(array)):
        key = array[step]
        j = step - 1    
        while j >= 0 and key < array[j]:
            array[j + 1] = array[j]
            j = j - 1
            count = count+1;
        array[j + 1] = key
    return count
```

```python
## Selection Sort
def selectionSort(itemsList, count):
    n = len(itemsList)
    for i in range(n - 1):
        minValueIndex = i
        for j in range(i + 1, n):
            if itemsList[j] < itemsList[minValueIndex]:
                minValueIndex = j
                count = count + 1
        if minValueIndex != i:
            temp = itemsList[i]
            itemsList[i] = itemsList[minValueIndex]
            itemsList[minValueIndex] = temp
            count = count + 1
    return count
```

```python
## Bubble sort

def bubbleSort(list, count):
    for passnum in range(len(list)-1, 0, -1):
      for i in range(passnum):
          if list[i] > list[i+1]:
              temp = list[i]
              list[i] = list[i+1]
              list[i+1] = temp
              count = count + 1
    return count
```

```python
## Merge Sort
def mergeSort(array):
    count = 0
    if len(array) > 1:
        r = len(array)//2
        L = array[:r]
        M = array[r:]
        mergeSort(L)
        mergeSort(M)
        i = j = k = 0
        while i < len(L) and j < len(M):
            if L[i] < M[j]:
                array[k] = L[i]
                i += 1
                count = count + 1
            else:
                array[k] = M[j]
                j += 1
            k += 1
        while i < len(L):
            array[k] = L[i]
            i += 1
            k += 1
            count = count + 1
        while j < len(M):
            array[k] = M[j]
            j += 1
            k += 1
            count = count + 1
    return count
```

```python
# Heap Sort
def heapify(arr, n, i):
    count = 0
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2

    if l < n and arr[i] < arr[l]:
        largest = l
        count = count + 1

    if r < n and arr[largest] < arr[r]:
        largest = r
        count = count + 1

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        count = count + 1
        heapify(arr, n, largest)
    
    return count

def heapSort(arr):
    count = 0
    n = len(arr)

    for i in range(n//2, -1, -1):
        count += heapify(arr, n, i)

    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        count += heapify(arr, i, 0)

    return count
```

```python
def partition(array, start, end):
    pivot = array[start]
    low = start + 1
    high = end
    count = 0

    while True:
        while low <= high and array[high] >= pivot:
            high = high - 1
            count += 1
        while low <= high and array[low] <= pivot:
            low = low + 1
            count += 1
        if low <= high:
            array[low], array[high] = array[high], array[low]
            count += 1
        else:
            break
    array[start], array[high] = array[high], array[start]
    return high, count

def quickSort(array, start, end):
    if start >= end:
        return
    p, count = partition(array, start, end)
    quickSort(array, start, p-1)
    quickSort(array, p+1, end)
    return count
```
